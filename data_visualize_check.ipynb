```python
# import packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# setup file paths and names
directory = '~/data/processed/'
acc_file = 'acc.csv'
bvp_file = 'bvp.csv'
eda_file = 'eda.csv'
hr_file = 'hr.csv'

# import eda data
eda_df = pd.read_csv(directory + eda_file, index_col='time (s)')

# import hr data
hr_df = pd.read_csv(directory + hr_file, index_col='time (s)')

# import bvp data
bvp_df = pd.read_csv(directory + bvp_file, index_col='time (s)')

# import acc data
acc_df = pd.read_csv(directory + acc_file, index_col='time (s)')

# participant names across dataframes are the same so use one as index to graph
participant_names = eda_df.columns.tolist()

# prompt user to select a participant number to show
show_plotnumber = int(input("enter a number between 0 and " + str(len(participant_names)) + ":"))
```

```python
# run this to see the plots of the participant you selected
plt.figure(figsize=(8, 20))
plt.subplot(4, 1, 1)
plt.plot(eda_df.index, eda_df[participant_names[show_plotnumber]])
plt.axvline(x=249, color='red', linestyle='dashed', linewidth=0.5)
plt.xlabel('Time (sec)')
plt.ylabel('Electrodermal Activity (microsiemens)')

plt.subplot(4, 1, 2)
plt.plot(hr_df.index, hr_df[participant_names[show_plotnumber]])
plt.axvline(x=249, color='red', linestyle='dashed', linewidth=0.5)
plt.xlabel('Time (sec)')
plt.ylabel('Heart Rate (bpm)')

plt.subplot(4, 1, 3)
plt.plot(bvp_df.index, bvp_df[participant_names[show_plotnumber]])
plt.axvline(x=249, color='red', linestyle='dashed', linewidth=0.5)
plt.xlabel('Time (sec)')
plt.ylabel('Photoplethysmograph data')

plt.subplot(4, 1, 4)
plt.plot(acc_df.index, acc_df[participant_names[show_plotnumber]])
plt.axvline(x=249, color='red', linestyle='dashed', linewidth=0.5)
plt.xlabel('Time (sec)')
plt.ylabel('G force')

plt.suptitle(participant_names[show_plotnumber])
plt.tight_layout()
plt.show()
```
